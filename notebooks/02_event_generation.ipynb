```python
# import libraries

import numpy as np
import pandas as pd
```

```python
# Set random seed
np.random.seed(42)
```

```python
# Load cleaned listings dataset
listings = pd.read_csv("../data/02-preprocessed/listings_clean.csv")
```

```python
# Preview cleaned listings dataset
print(listings.shape)
listings.head()
```

```python
# Define sample size for synthetic event generation
sample_size = 30000

# Sample listings with replacement to simulate repeated exposure across sessions
events_base = listings.sample(n=sample_size, replace=True).reset_index(drop=True)
```

```python
events_base
```

```python
# Create identifiers
n_events = len(events_base)

events_base["event_id"] = np.arange(1, n_events + 1)
events_base["user_id"] = np.random.randint(10000, 99999, size=n_events)
events_base["session_id"] = np.random.randint(100000, 999999, size=n_events)
```

```python
events_base
```

```python
# Create event timestamps
start_date = pd.to_datetime("2025-01-01")

events_base["event_time"] = start_date + pd.to_timedelta(
    np.random.randint(0, 90 * 24 * 60, size=n_events),
    unit="m"
)
```

```python
events_base
```

```python
# Create device type
events_base["device_type"] = np.random.choice(
    ["mobile", "desktop", "tablet"],
    size=n_events,
    p=[0.6, 0.3, 0.1]
)

# Create traffic source
events_base["traffic_source"] = np.random.choice(
    ["organic", "paid", "email", "direct"],
    size=n_events,
    p=[0.5, 0.2, 0.1, 0.2]
)
```

```python
# Create impression flag
events_base["impression"] = 1
```

```python
# Create click score
click_score = (
    0.8 * events_base["availability_ratio"]
    + 0.6 * (events_base["room_type"] == "Entire home/apt").astype(int)
    + 0.4 * (events_base["city"] == "Manhattan").astype(int)
    + 0.3 * np.log1p(events_base["demand_score"])
    - 0.15 * events_base["price_log"]
)

# Convert score to probability
click_prob = 1 / (1 + np.exp(-click_score))
click_prob = np.clip(click_prob, 0.05, 0.95)
```

```python
# Generate click flag
events_base["click"] = np.random.binomial(1, click_prob)
```

```python
# Create detail view probability
detail_prob = np.where(events_base["click"] == 1, 0.75, 0.0)
```

```python
# Generate detail view flag
events_base["detail_view"] = np.random.binomial(1, detail_prob)
```

```python
# Create save score
save_score = (
    0.5 * np.log1p(events_base["demand_score"])
    + 0.4 * (events_base["room_type"] == "Entire home/apt").astype(int)
    - 0.1 * events_base["price_log"]
)

# Convert score to probability
save_prob = 1 / (1 + np.exp(-save_score))
save_prob = np.clip(save_prob, 0.01, 0.85)

# Only allow saves after detail views
save_prob = np.where(events_base["detail_view"] == 1, save_prob, 0.0)
```

```python
# Generate save flag
events_base["save"] = np.random.binomial(1, save_prob)
```

```python
# Create contact score
contact_score = (
    0.7 * events_base["save"]
    + 0.5 * events_base["detail_view"]
    + 0.3 * np.log1p(events_base["demand_score"])
    + 0.2 * (events_base["city"] == "Manhattan").astype(int)
    - 0.2 * events_base["price_log"]
)

# Convert score to probability
contact_prob = 1 / (1 + np.exp(-contact_score))
contact_prob = np.clip(contact_prob, 0.01, 0.75)

# Only allow contacts after detail views
contact_prob = np.where(events_base["detail_view"] == 1, contact_prob, 0.0)
```

```python
# Generate contact flag
events_base["contact"] = np.random.binomial(1, contact_prob)
```

```python
# Select final event columns
events = events_base[
    [
        "event_id",
        "user_id",
        "session_id",
        "listing_id",
        "event_time",
        "device_type",
        "traffic_source",
        "room_type",
        "city",
        "price",
        "price_log",
        "demand_score",
        "availability_ratio",
        "impression",
        "click",
        "detail_view",
        "save",
        "contact"
    ]
].copy()
```

```python
# Preview events dataset
print(events.shape)
events.head()
```

```python
# Check funnel rates
funnel_summary = pd.DataFrame({
    "metric": [
        "impression_rate",
        "click_rate",
        "detail_view_rate",
        "save_rate",
        "contact_rate"
    ],
    "value": [
        events["impression"].mean(),
        events["click"].mean(),
        events["detail_view"].mean(),
        events["save"].mean(),
        events["contact"].mean()
    ]
})

funnel_summary
```

```python
# Save events dataset
events.to_csv("../data/01-raw/events.csv", index=False)
```
